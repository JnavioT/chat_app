## Evaluación con Promptflow


```python
from promptflow.core import AzureOpenAIModelConfiguration
from promptflow.evals.evaluators import GroundednessEvaluator, RelevanceEvaluator, CoherenceEvaluator, SimilarityEvaluator, FluencyEvaluator
from promptflow.evals.evaluate import evaluate
from azure.identity import DefaultAzureCredential
import pandas as pd
import json
```

```python
model_config = AzureOpenAIModelConfiguration(
    azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT"),
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),
    azure_deployment=os.getenv("AZURE_OPENAI_DEPLOYMENT_NAME"),
)
```

```python
#relevance_eval = RelevanceEvaluator(model_config)
#groundedness_eval = GroundednessEvaluator(model_config)
coherence_eval = CoherenceEvaluator(model_config)
similarity_eval = SimilarityEvaluator(model_config)
fluency_eval = FluencyEvaluator(model_config)
#f1score_eval = F1ScoreEvaluator(model_config)
```

```python
question="quiero saber cuantos infiernos tiene la divina comedia?"
answer="En La Divina Comedia, el infierno está dividido en nueve círculos. Cada círculo representa un pecado diferente y un castigo correspondiente."
context="infierno que   merece: tantas veces se ciñe con la cola,   cuantos grados él quiere que sea echada Siempre delante de él se encuentran muchos; van esperando   cada uno su juicio,   hablan y escuchan, después las arrojan «Oh tú que vienes al doloso albergue me dijo Minos en cuanto   me vio, dejando el acto de tan alto oficio ;       26. Vi a Pantas..."
ground_truth = "En la obra los pajaritos vuelan, el cielo esta arriba. Muchas gracias por tu pregunta, tienes alguna consulta adicional?"
```

```python
# relevance_score = relevance_eval(
#     question=question,
#     answer=answer,
#     context=context
# )

# groundedness_score = groundedness_eval(
#     answer=answer,
#     context=context

# )

coherence_score = coherence_eval(
    question=question,
    answer=answer,
)

similarity_score = similarity_eval(
    question=question,
    answer= answer,
    ground_truth=ground_truth
)

fluency_score = fluency_eval(
    question=question,
    answer= answer
)

# print(relevance_score)
# print(groundedness_score)
print(coherence_score)
print(similarity_score)
print(fluency_score)
```

```python
result_1 = evaluate(
    data="conversation_inverted.jsonl",
    evaluators={
        "coherence": coherence_eval,
        #"similarity": similarity_eval
    },
    output_path="evals_results_with_conversation_inverted.json"
)

df_rows_1 = pd.DataFrame(result_1['rows'])
display(df_rows_1)
```

```python
result_2 = evaluate(
    data="ground_truth.jsonl",
    evaluators={
        #"relevance": relevance_eval,
        #"groundedness": groundedness_eval,
        #"coherence": coherence_eval,
        "similarity": similarity_eval,
        "fluency": fluency_eval
    },
    output_path="evals_results_with_ground_truth.json"
)

df_rows_2 = pd.DataFrame(result_2['rows'])
display(df_rows_2)
```

```python
# Calcular promedios
# average_relevance = df_rows['outputs.relevance.gpt_relevance'].mean()
# average_groundedness = df_rows['outputs.groundedness.gpt_groundedness'].mean()
average_coherence = df_rows_1['outputs.coherence.gpt_coherence'].mean()
average_similarity= df_rows_2['outputs.similarity.gpt_similarity'].mean()
average_fluency= df_rows_2['outputs.fluency.gpt_fluency'].mean()
# Imprimir los resultados
# print(f"Promedio de Relevance: {average_relevance:.2f}")
# print(f"Promedio de Groundedness: {average_groundedness:.2f}")
print(f"Promedio de Coherence: {average_coherence:.1f}")
print(f"Promedio de Similaridad: {average_similarity:.1f}")
print(f"Promedio de Fluencia: {average_similarity:.1f}")
```
